```python
# Write your implementation here
# TODO: Write the words (including spellings, frequencies and neighbours) from the lexicon to a text file
class AVLTree:

  def insert(self, root, key):
    # remove non-alphabetic characters from key and convert to lowercase
    cleanedKey = ''.join([ch for ch in key if ch.isalpha()]).lower()

    # binary search
    if not root:
      return Word(cleanedKey)
    elif cleanedKey < root.word:
      root.left = self.insert(root.left, cleanedKey)
    elif cleanedKey > root.word:
      root.right = self.insert(root.right, cleanedKey)
    else:  # if the word is found, increase its frequency
      root.increaseFrequency()
      return root

    # update the height of the ancestor node
    root.height = 1 + max(self.getHeight(root.left), self.getHeight(root.right))

    # get the balance
    balance = self.getBalance(root)

    # if the node is unbalanced, then try out the 4 cases
    # case 1: left left
    if balance > 1 and cleanedKey < root.left.word:
      return self.rightRotate(root)

    # case 2: right right
    if balance < -1 and cleanedKey > root.right.word:
      return self.leftRotate(root)

    # case 3: left right
    if balance > 1 and cleanedKey > root.left.word:
      root.left = self.leftRotate(root.left)
      return self.rightRotate(root)

    # case 4: right left
    if balance < -1 and cleanedKey < root.right.word:
      root.right = self.rightRotate(root.right)
      return self.leftRotate(root)

    return root

  def leftRotate(self, z):
    y = z.right
    T2 = y.left

    # perform rotation
    y.left = z
    z.right = T2

    # update heights
    z.height = 1 + max(self.getHeight(z.left), self.getHeight(z.right))
    y.height = 1 + max(self.getHeight(y.left), self.getHeight(y.right))

    # get the new root
    return y

  def rightRotate(self, z):
    y = z.left
    T3 = y.right

    # perform rotation
    y.right = z
    z.left = T3

    # update heights
    z.height = 1 + max(self.getHeight(z.left), self.getHeight(z.right))
    y.height = 1 + max(self.getHeight(y.left), self.getHeight(y.right))

    # get the new root
    return y

  def getHeight(self, root): # get the height of the avl tree
    if not root:
      return 0
    return root.height

  def getBalance(self, root): # get the avl tree balance
    if not root:
      return 0
    return self.getHeight(root.left) - self.getHeight(root.right)

  def findWord(self, root, word): # progressively go through the tree to find a word, til it is returned, pretty sure i didnt actually use this
    if root is None or root.word == word:
      return root
    elif word < root.word:
      return self.findWord(root.left, word)
    else:
      return self.findWord(root.right, word)

  def getAllWords(self, root): # gets all words from the lexicon and puts into an array
    result = []
    def inOrderTraversal(root):
      if root:
        inOrderTraversal(root.left)
        result.append(root)
        inOrderTraversal(root.right)
    inOrderTraversal(root)
    return result

  def preOrder(self, root): # data in avl tree traversal order
    if not root:
      return
    print(f"{root.word} {root.frequency} {root.neighbours}")
    self.preOrder(root.left)
    self.preOrder(root.right)

  def inOrder(self, root): # data in actual order
    if not root:
      return
    self.inOrder(root.left)
    print(f"{root.word} {root.frequency} {root.neighbours}")
    self.inOrder(root.right)

  def printInOrder(self, root): # variation on pre order used for writing to the output file mainly
    if not root:
      return ""
    return self.printInOrder(root.left) + f"{root.word} {root.frequency} {root.neighbours}\n" + self.printInOrder(root.right)

# testing
lexicon = AVLTree()
root = None

#with open("s1.txt", "r") as file:
  # split file into words, then add each to the lexicon
  #for line in file:
    #words = line.split()
    #for word in words:
      #if word != ' ':
        #root = lexicon.insert(root, word)

lexicon.inOrder(root)

# Print the list of words to the console
with open('testoutput.txt', 'w') as file:
  file.write(lexicon.printInOrder(root))
```

```python
def add_neighbours(lexicon, root):
  # temporary dictionary to store words by their length, makes it much faster to search for neighbours if you only look through words with the same length
  lengthTemp = {}

  # add words to dictionary grouped by length
  words = lexicon.getAllWords(root)
  for word in words:
    curLength = len(word.word) # current length of words we are looking at
    if curLength not in lengthTemp:
      lengthTemp[curLength] = []
    lengthTemp[curLength].append(word)

  # Iterate over each group of words with the same length, largely the same as previous implementation
  for length in lengthTemp:
    wordsWithLength = lengthTemp[length]
    for a in wordsWithLength:
      for b in wordsWithLength:
        if a.word != b.word:
          letterDiff = 0
          for i in range(len(a.word)):
            if a.word[i] != b.word[i]:
              letterDiff += 1
          if letterDiff == 1:
            a.neighbours.append(b.word)


# Call your add_neighbours() function with your lexicon
add_neighbours(lexicon, root)
lexicon.inOrder(root)
```
